Tagged Corpora:In addition to the plaintext corpora,NLTK also contains a wide varitey of annoted text.


```python
from nltk.corpus import brown
```

```python
print(brown.words())
```

```python
print(brown.tagged_words())
```

```python
print(brown.sents())
print(brown.tagged_sents())
```

```python
print(brown.paras(categories='reviews'))
print(brown.tagged_paras(categories='reviews'))
```
